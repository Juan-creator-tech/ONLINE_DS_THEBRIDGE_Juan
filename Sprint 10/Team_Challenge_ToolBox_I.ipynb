### Funcion: describe_df

Esta funcin debe recibir como argumento un dataframe y debe devolver una dataframe como el de la imagen (no el de la imagen). Es decir un dataframe que tenga una columna por cada columan del dataframe original y como filas, los tipos de las columnas, el tanto por ciento de valores nulos o missings, los valores nicos y el porcentaje de cardinalidad. 

La figura muestra el resultado esperado de llamar a la funcin pedida con el dataset del titanic:

    

````python
def describe_df(df):
    """
    Genera un resumen descriptivo de un DataFrame dado.

    :param df: pd.DataFrame - DataFrame de entrada
    :return: pd.DataFrame - DataFrame descriptivo
    """
    resumen = {
        "Tipo de dato": df.dtypes,
        "% Nulos": df.isnull().mean() * 100,
        "Valores nicos": df.nunique(),
        "% Cardinalidad": df.nunique() / len(df) * 100
    }

    # Convertir el diccionario a DataFrame y redondear los valores numricos
    resumen_df = pd.DataFrame(resumen).T
    resumen_df = resumen_df.round(2)

    return resumen_df
````

### Funcion: tipifica_variables

Esta funcin debe recibir como argumento un dataframe, un entero (`umbral_categoria`) y un float (`umbral_continua`). La funcin debe devolver un dataframe con dos columnas "nombre_variable", "tipo_sugerido" que tendr tantas filas como columnas el dataframe. En cada fila ir el nombre de una de las columnas y una sugerencia del tipo de variable. Esta sugerencia se har siguiendo las siguientes pautas:
+ Si la cardinalidad es 2, asignara "Binaria"
+ Si la cardinalidad es menor que `umbral_categoria` asignara "Categrica"
+ Si la cardinalidad es mayor o igual que `umbral_categoria`, entonces entra en juego el tercer argumento:
    * Si adems el porcentaje de cardinalidad es superior o igual a `umbral_continua`, asigna "Numerica Continua"
    * En caso contrario, asigna "Numerica Discreta"


````python
def tipifica_variables(df, umbral_categoria, umbral_continua):
    """
    Tipifica las variables de un DataFrame basado en su cardinalidad y porcentaje de cardinalidad.

    :param df: pd.DataFrame - DataFrame de entrada
    :param umbral_categoria: int - Umbral para considerar una variable categrica
    :param umbral_continua: float - Umbral para considerar una variable continua
    :return: pd.DataFrame - DataFrame con las columnas "nombre_variable" y "tipo_sugerido"
    """
    tipos = []

    for col in df.columns:
        cardinalidad = df[col].nunique()
        porcentaje_cardinalidad = (cardinalidad / len(df)) * 100

        if cardinalidad == 2:
            tipo = "Binaria"
        elif cardinalidad < umbral_categoria:
            tipo = "Categrica"
        else:
            if porcentaje_cardinalidad >= umbral_continua:
                tipo = "Numerica Continua"
            else:
                tipo = "Numerica Discreta"

        tipos.append({"nombre_variable": col, "tipo_sugerido": tipo})

    return pd.DataFrame(tipos)
````

### Funcion: get_features_num_regression

Esta funcin recibe como argumentos un dataframe, el nombre de una de las columnas del mismo (argumento 'target_col'), que debera ser el target de un hipottico modelo de regresin, es decir debe ser una variable numrica continua o discreta pero con alta cardinalidad, adems de un argumento 'umbral_corr', de tipo float que debe estar entre 0 y 1 y una variable float "pvalue" cuyo valor debe ser por defecto "None".

La funcin debe devolver una lista con las columnas numricas del dataframe cuya correlacin con la columna designada por "target_col" sea superior en valor absoluto al valor dado por "umbral_corr". Adems si la variable "pvalue" es distinta de None, slo devolvera las columnas numricas cuya correlacin supere el valor indicado y adems supere el test de hiptesis con significacin mayor o igual a 1-pvalue.

La funcin debe hacer todas las comprobaciones necesarias para no dar error como consecuecia de los valores de entrada. Es decir har un check de los valores asignados a los argumentos de entrada y si estos no son adecuados debe retornar None y printar por pantalla la razn de este comportamiento. Ojo entre las comprobaciones debe estar que "target_col" hace referencia a una variable numrica continua del dataframe.

````python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

def get_features_num_regression(df, target_col, umbral_corr, pvalue=None):
    """
    Selecciona columnas numricas con alta correlacin con el target para regresin.

    :param df: pd.DataFrame - DataFrame de entrada.
    :param target_col: str - Columna objetivo.
    :param umbral_corr: float - Umbral mnimo de correlacin en valor absoluto.
    :param pvalue: float - Valor p para la significancia estadstica (opcional).
    :return: list - Lista de nombres de columnas que cumplen los criterios.
    """
    # Validaciones de entrada
    if not isinstance(df, pd.DataFrame):
        print(" El argumento 'df' debe ser un DataFrame.")
        return None

    if target_col not in df.columns:
        print(f" La columna objetivo '{target_col}' no est en el DataFrame.")
        return None

    if not pd.api.types.is_numeric_dtype(df[target_col]):
        print(f" La columna objetivo '{target_col}' debe ser numrica y continua.")
        return None

    if not (0 <= umbral_corr <= 1):
        print(" El argumento 'umbral_corr' debe estar entre 0 y 1.")
        return None

    if pvalue is not None and not (0 <= pvalue <= 1):
        print(" El argumento 'pvalue' debe estar entre 0 y 1 o ser None.")
        return None

    # Seleccin de columnas numricas
    num_cols = df.select_dtypes(include=['number']).columns.tolist()
    if target_col in num_cols:
        num_cols.remove(target_col)

    if not num_cols:
        print(" No se encontraron columnas numricas en el DataFrame aparte de la columna objetivo.")
        return []

    seleccionadas = []

    for col in num_cols:
        try:
            correlacion = df[[col, target_col]].dropna()
            corr, p = pearsonr(correlacion[col], correlacion[target_col])
            if abs(corr) >= umbral_corr:
                if pvalue is None or p <= (1 - pvalue):
                    seleccionadas.append(col)
        except Exception as e:
            print(f" Error al calcular la correlacin para la columna '{col}': {e}")

    if not seleccionadas:
        print(" No se encontraron columnas que cumplan con los criterios especificados.")
        return []

    print(f" Columnas seleccionadas: {seleccionadas}")
    return seleccionadas
````

````python
import seaborn as sns

# Carga de un dataset de ejemplo
df = sns.load_dataset('iris')
df['target'] = df['sepal_length'] * 1.5  # Agregar una columna objetivo simulada

# Llamada a la funcin
resultado = get_features_num_regression(df, target_col='target', umbral_corr=0.5, pvalue=0.05)
print("Resultado:", resultado)
````

### Funcion: plot_features_num_regression

Si la lista no est vaca, la funcin pintar una pairplot del dataframe considerando la columna designada por "target_col" y aquellas incluidas en "column" que cumplan que su correlacin con "target_col" es superior en valor absoluto a "umbral_corr", y que, en el caso de ser pvalue diferente de "None", adems cumplan el test de correlacin para el nivel 1-pvalue de significacin estadstica. La funcin devolver los valores de "columns" que cumplan con las condiciones anteriores. 

EXTRA: Se valorar adicionalmente el hecho de que si la lista de columnas a pintar es grande se pinten varios pairplot con un mximo de cinco columnas en cada pairplot (siendo siempre una de ellas la indicada por "target_col")

De igual manera que en la funcin descrita anteriormente deber hacer un check de los valores de entrada y comportarse como se describe en el ltimo prrafo de la funcin `get_features_num_regresion`

````python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt

def plot_features_num_regression(df, target_col="", columns=[], umbral_corr=0, pvalue=None):
    """
    Genera pairplots para las columnas numricas seleccionadas en base a su correlacin con una columna objetivo.

    :param df: pd.DataFrame - DataFrame de entrada
    :param target_col: str - Columna objetivo
    :param columns: list - Lista de columnas a considerar
    :param umbral_corr: float - Umbral mnimo de correlacin en valor absoluto
    :param pvalue: float - Valor p para la significancia estadstica (opcional)
    :return: list - Lista de columnas seleccionadas
    """
    # Validaciones de entrada
    if not isinstance(df, pd.DataFrame):
        print(" El argumento 'df' debe ser un DataFrame.")
        return None

    if target_col not in df.columns:
        print(f" La columna objetivo '{target_col}' no est en el DataFrame.")
        return None

    if not pd.api.types.is_numeric_dtype(df[target_col]):
        print(f" La columna objetivo '{target_col}' debe ser numrica y continua.")
        return None

    if not (0 <= umbral_corr <= 1):
        print(" El argumento 'umbral_corr' debe estar entre 0 y 1.")
        return None

    if pvalue is not None and not (0 <= pvalue <= 1):
        print(" El argumento 'pvalue' debe estar entre 0 y 1 o ser None.")
        return None

    # Si la lista de columnas est vaca, seleccionamos todas las numricas menos la columna objetivo
    if not columns:
        columns = df.select_dtypes(include=['number']).columns.tolist()
        if target_col in columns:
            columns.remove(target_col)

    # Validar columnas
    if not all(col in df.columns for col in columns):
        print(" Algunas columnas especificadas no estn en el DataFrame.")
        return None

    if not all(pd.api.types.is_numeric_dtype(df[col]) for col in columns):
        print(" Todas las columnas seleccionadas deben ser numricas.")
        return None

    seleccionadas = []

    for col in columns:
        try:
            correlacion = df[[col, target_col]].dropna()
            corr, p = pearsonr(correlacion[col], correlacion[target_col])
            if abs(corr) >= umbral_corr:
                if pvalue is None or p <= (1 - pvalue):
                    seleccionadas.append(col)
        except Exception as e:
            print(f" Error al calcular la correlacin para la columna '{col}': {e}")

    if not seleccionadas:
        print(" No se encontraron columnas que cumplan con los criterios especificados.")
        return []

    # Generar pairplots
    max_cols_per_plot = 5
    print(f" Columnas seleccionadas para pairplot: {seleccionadas}")
    for i in range(0, len(seleccionadas), max_cols_per_plot - 1):
        subset = seleccionadas[i:i + max_cols_per_plot - 1] + [target_col]
        try:
            sns.pairplot(df[subset].dropna())
            plt.show()
        except Exception as e:
            print(f" Error al generar el pairplot para el subconjunto {subset}: {e}")

    return seleccionadas
````

````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, f_oneway

def describe_df(df):
    """
    Genera un resumen descriptivo de un DataFrame dado.

    :param df: pd.DataFrame - DataFrame de entrada
    :return: pd.DataFrame - DataFrame descriptivo
    """
    resumen = {
        "Tipo de dato": df.dtypes,
        "% Nulos": df.isnull().mean() * 100,
        "Valores nicos": df.nunique(),
        "% Cardinalidad": df.nunique() / len(df) * 100
    }

    # Convertir el diccionario a DataFrame y redondear los valores numricos
    resumen_df = pd.DataFrame(resumen).T
    resumen_df = resumen_df.round(2)

    return resumen_df

def tipifica_variables(df, umbral_categoria, umbral_continua):
    """
    Tipifica las variables de un DataFrame basado en su cardinalidad y porcentaje de cardinalidad.

    :param df: pd.DataFrame - DataFrame de entrada
    :param umbral_categoria: int - Umbral para considerar una variable categrica
    :param umbral_continua: float - Umbral para considerar una variable continua
    :return: pd.DataFrame - DataFrame con las columnas "nombre_variable" y "tipo_sugerido"
    """
    tipos = []

    for col in df.columns:
        cardinalidad = df[col].nunique()
        porcentaje_cardinalidad = (cardinalidad / len(df)) * 100

        if cardinalidad == 2:
            tipo = "Binaria"
        elif cardinalidad < umbral_categoria:
            tipo = "Categrica"
        else:
            if porcentaje_cardinalidad >= umbral_continua:
                tipo = "Numerica Continua"
            else:
                tipo = "Numerica Discreta"

        tipos.append({"nombre_variable": col, "tipo_sugerido": tipo})

    return pd.DataFrame(tipos)

def get_features_num_regression(df, target_col, umbral_corr, pvalue=None):
    """
    Selecciona columnas numricas con alta correlacin con el target para regresin.

    :param df: pd.DataFrame - DataFrame de entrada
    :param target_col: str - Columna objetivo
    :param umbral_corr: float - Umbral mnimo de correlacin en valor absoluto
    :param pvalue: float - Valor p para la significancia estadstica (opcional)
    :return: list - Lista de nombres de columnas que cumplen los criterios
    """
    # Validaciones de entrada
    if not isinstance(df, pd.DataFrame):
        print("El argumento 'df' debe ser un DataFrame.")
        return None

    if target_col not in df.columns:
        print(f"La columna objetivo '{target_col}' no est en el DataFrame.")
        return None

    if not pd.api.types.is_numeric_dtype(df[target_col]):
        print(f"La columna objetivo '{target_col}' debe ser numrica.")
        return None

    if not (0 <= umbral_corr <= 1):
        print("El argumento 'umbral_corr' debe estar entre 0 y 1.")
        return None

    if pvalue is not None and not (0 <= pvalue <= 1):
        print("El argumento 'pvalue' debe estar entre 0 y 1 o ser None.")
        return None

    # Seleccin de columnas numricas
    num_cols = df.select_dtypes(include=['number']).columns.tolist()
    num_cols.remove(target_col)

    seleccionadas = []

    for col in num_cols:
        correlacion = df[[col, target_col]].dropna()
        corr, p = pearsonr(correlacion[col], correlacion[target_col])

        if abs(corr) >= umbral_corr:
            if pvalue is None or p <= (1 - pvalue):
                seleccionadas.append(col)

    return seleccionadas

def plot_features_num_regression(df, target_col="", columns=[], umbral_corr=0, pvalue=None):
    """
    Genera pairplots para las columnas numricas seleccionadas en base a su correlacin con una columna objetivo.

    :param df: pd.DataFrame - DataFrame de entrada
    :param target_col: str - Columna objetivo
    :param columns: list - Lista de columnas a considerar
    :param umbral_corr: float - Umbral mnimo de correlacin en valor absoluto
    :param pvalue: float - Valor p para la significancia estadstica (opcional)
    :return: list - Lista de columnas seleccionadas
    """
    # Validaciones de entrada
    if not isinstance(df, pd.DataFrame):
        print("El argumento 'df' debe ser un DataFrame.")
        return None

    if target_col not in df.columns:
        print(f"La columna objetivo '{target_col}' no est en el DataFrame.")
        return None

    if not pd.api.types.is_numeric_dtype(df[target_col]):
        print(f"La columna objetivo '{target_col}' debe ser numrica.")
        return None

    if not (0 <= umbral_corr <= 1):
        print("El argumento 'umbral_corr' debe estar entre 0 y 1.")
        return None

    if pvalue is not None and not (0 <= pvalue <= 1):
        print("El argumento 'pvalue' debe estar entre 0 y 1 o ser None.")
        return None

    # Si la lista de columnas est vaca, seleccionamos todas las numricas
    if not columns:
        columns = df.select_dtypes(include=['number']).columns.tolist()
        columns.remove(target_col)

    seleccionadas = []

    for col in columns:
        correlacion = df[[col, target_col]].dropna()
        corr, p = pearsonr(correlacion[col], correlacion[target_col])

        if abs(corr) >= umbral_corr:
            if pvalue is None or p <= (1 - pvalue):
                seleccionadas.append(col)

    # Generar pairplots
    max_cols_per_plot = 5
    for i in range(0, len(seleccionadas), max_cols_per_plot - 1):
        subset = seleccionadas[i:i + max_cols_per_plot - 1] + [target_col]
        sns.pairplot(df[subset].dropna())
        plt.show()

    return seleccionadas

def get_features_cat_regression(df, target_col, pvalue=0.05):
    """
    Selecciona columnas categricas relacionadas con el target para regresin.

    :param df: pd.DataFrame - DataFrame de entrada
    :param target_col: str - Columna objetivo
    :param pvalue: float - Valor p para la significancia estadstica
    :return: list - Lista de nombres de columnas categricas que cumplen los criterios
    """
    # Validaciones de entrada
    if not isinstance(df, pd.DataFrame):
        print("El argumento 'df' debe ser un DataFrame.")
        return None

    if target_col not in df.columns:
        print(f"La columna objetivo '{target_col}' no est en el DataFrame.")
        return None

    if not pd.api.types.is_numeric_dtype(df[target_col]):
        print(f"La columna objetivo '{target_col}' debe ser numrica.")
        return None

    if not (0 <= pvalue <= 1):
        print("El argumento 'pvalue' debe estar entre 0 y 1.")
        return None

    # Seleccin de columnas categricas
    cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
    seleccionadas = []

    for col in cat_cols:
        grupos = [df[target_col][df[col] == val].dropna() for val in df[col].unique() if pd.notnull(val)]
        if len(grupos) > 1:
            stat, p = f_oneway(*grupos)
            if p <= pvalue:
                seleccionadas.append(col)

    return seleccionadas

def plot_features_cat_regression(df, target_col="", columns=[], pvalue=0.05, with_individual_plot=False):
    """
    Genera histogramas agrupados para variables categricas seleccionadas relacionadas con una columna objetivo.

    :param df: pd.DataFrame - DataFrame de entrada
    :param target_col: str - Columna objetivo
    :param columns: list - Lista de columnas categricas a considerar
    :param pvalue: float - Valor p para la significancia estadstica
    :param with_individual_plot: bool - Si True, genera grficos individuales para cada categora
    :return: list - Lista de columnas seleccionadas
    """
    # Validaciones de entrada
    if not isinstance(df, pd.DataFrame):
        print("El argumento 'df' debe ser un DataFrame.")
        return None

    if target_col not in df.columns:
        print(f"La columna objetivo '{target_col}' no est en el DataFrame.")
        return None

    if not pd.api.types.is_numeric_dtype(df[target_col]):
        print(f"La columna objetivo '{target_col}' debe ser numrica.")
        return None

    if not (0 <= pvalue <= 1):
        print("El argumento 'pvalue' debe estar entre 0 y 1.")
        return None

    # Si la lista de columnas est vaca, seleccionamos todas las categricas
    if not columns:
        columns = df.select_dtypes(include=['object', 'category']).columns.tolist()

    seleccionadas = []

    for col in columns:
        grupos = [df[target_col][df[col] == val].dropna() for val in df[col].unique() if pd.notnull(val)]
        if len(grupos) > 1:
            stat, p = f_oneway(*grupos)
            if p <= pvalue:
                seleccionadas.append(col)

    # Generar grficos
    for col in seleccionadas:
        if with_individual_plot:
            for val in df[col].dropna().unique():
                subset = df[df[col] == val]
                sns.histplot(subset[target_col], kde=True, label=f"{col} = {val}")
                plt.legend()
                plt.title(f"Distribucin de {target_col} para {col} = {val}")
                plt.show()
        else:
            sns.histplot(data=df, x=target_col, hue=col, kde=True, multiple="stack")
            plt.title(f"Distribucin de {target_col} agrupada por {col}")
            plt.show()

    return seleccionadas

````

````python
seleccionadas_num = get_features_num_regression(df, target_col="survived", umbral_corr=0.5, pvalue=0.05)
seleccionadas_num
````

````python
seleccionadas_cat = get_features_cat_regression(df, target_col="survived", pvalue=0.05)
seleccionadas_cat
````

````python
columnas_plot = plot_features_num_regression(df, target_col="survived", columns=[], umbral_corr=0.5, pvalue=0.05)
columnas_plot
````

````python
plot_features_cat_regression(df, target_col="survived", columns=seleccionadas_cat, pvalue=0.05, with_individual_plot=True)
````
